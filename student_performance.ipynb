```python
# get mean, median, minimum, maximum for g1,g2,g3
described_value = dataframe.describe()
print(described_value[["G1","G2","G3"]].loc[["mean","50%","min","max"]])
```

```python
# Question answer phase using code: 
```

```python
terms = ["G1","G2","G3"]
def get_high_low_term_mark(terms, dat="highest"):
    if dat=="highest":
        score = float("-inf")
    else:
        score = float("inf")
    term = ""
    for i in terms:
        marks = described_value[i]["mean"]
        if (dat == "highest" and marks > score) | (dat == "lowest" and marks < score):
            score = marks
            term = i      
    print(f"The term with {dat} mark is: {term} with {score} score")
```

```python
# Which term has the highest average marks?
get_high_low_term_mark(terms, "highest")
```

```python
# Which term has the lowest average marks?
get_high_low_term_mark(terms, "lowest")
```
